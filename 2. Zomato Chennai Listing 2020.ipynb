```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.io as pio

# Set default renderer to SVG for Plotly
pio.renderers.default = "svg"
```

```python
# Load the dataset
raw_df = pd.read_csv('Zomato Chennai Listing 2020.csv')

```

```python
print("First few rows of the dataset:")
print(raw_df.head())
```

```python
print("\nShape of the dataset:", raw_df.shape)
```

```python
print("\nInformation about the dataset:")
print(raw_df.info())
```

```python
print("\nSummary statistics of the dataset:")
print(raw_df.describe())
```

```python
print("\nMissing values:")
print(raw_df.isnull().sum())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(raw_df['Price for 2'], bins=20, kde=True)
plt.title('Histogram of Price for 2')
plt.xlabel('Price for 2')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
cuisine_counts = raw_df['Cuisine'].value_counts().head(10)
sns.barplot(x=cuisine_counts.index, y=cuisine_counts.values)
plt.title('Top 10 Cuisine Types')
plt.xlabel('Cuisine Type')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
location_counts = raw_df['Location'].value_counts()
fig = px.pie(location_counts, values=location_counts.values, names=location_counts.index, title='Location Distribution')
fig.show()
```
